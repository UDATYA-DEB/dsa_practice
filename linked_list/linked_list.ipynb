```python
class Node:
    def __init__(self, data):
        self.next = None
        self.data = data

def print_nodes(head):
    ptr = head
    
    while ptr is not None:
        print(ptr.data, end='->')
        ptr = ptr.next
    print("None")
```

```python
def generate_llist():
    arr = range(1, 6)

    head = Node(arr[0])
    ptr = head

    for i in arr[1:]:
        node = Node(i)
        ptr.next = node
        ptr = node
    
    return head

head = generate_llist()
print_nodes(head)
```

```python
#   find element in linked_list return index, if not found return -1
#   https://www.geeksforgeeks.org/problems/search-in-linked-list-1664434326/1

def find_element(head, val):
    ptr = head
    
    if ptr is None:
        return -1
    
    index = 0
    while ptr is not None:
        if ptr.data == val:
            return index
        index += 1
        ptr = ptr.next
    
    if ptr is None:
        return -1
```

```python
head = generate_llist()
print_nodes(head)
find_element(head, 10)
```

```python
#   delete head of the linked_list
#   https://www.geeksforgeeks.org/problems/delete-head-of-linked-list/1

def delete_head(head):
    if head is None:
        return None
    return head.next

head = generate_llist()
print_nodes(head)
new_head = delete_head(head)
print_nodes(new_head)
```

```python
#   delete tail of linked_list
#   https://www.geeksforgeeks.org/problems/deletion-at-the-end-of-a-linked-list/1

def delete_tail(head):
    if head is None or head.next is None:
        return None
    
    ptr = head
    
    while ptr.next.next is not None:
        ptr = ptr.next
    
    ptr.next = None
    
    return head

head = generate_llist()
print_nodes(head)
new_head = delete_tail(head)
print_nodes(new_head)
```

```python
#   delete kth element (0th index based)
#   https://www.geeksforgeeks.org/problems/delete-a-node-in-single-linked-list/1

def delete_node(head, index):
    if index == 0:
        if head is None or head.next is None:
            return None
        else:
            return head.next
    
    curr = head.next
    prev = head
    calc_index = 1
    
    while curr is not None:
        if calc_index == index:
            prev.next = curr.next
            del curr
            break
        
        prev = curr
        curr = curr.next
        calc_index += 1
    
    return head

def opt_delete_node(head, index):
    if index == 0:
        if head is None or head.next is None:
            return None
        else:
            new_head = head.next
            del head
            return new_head
    
    ptr = head
    index -= 1
    
    while ptr.next is not None:
        if index == 0:
            temp = ptr.next
            ptr.next = temp.next
            del temp
            break
        
        ptr = ptr.next
        index -= 1
    
    return head

def delete_node_el(head, element):
    if head is None:
        return None
    if head.data == element:
        return head.next
    
    curr = head.next
    prev = head
    
    while curr is not None:
        if curr.data == element:
            prev.next = curr.next
            del curr
            break
        
        prev = curr
        curr = curr.next
    
    return head

head = generate_llist()
print_nodes(head)
# new_head = delete_node(head, 3)
# new_head = opt_delete_node(head, 5)
new_head = delete_node_el(head, 6)
print_nodes(new_head)

    
```
