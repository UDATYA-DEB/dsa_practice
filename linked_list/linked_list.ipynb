```python
class Node:
    def __init__(self, data):
        self.next = None
        self.data = data

def print_nodes(head):
    ptr = head
    
    while ptr is not None:
        print(ptr.data, end='->')
        ptr = ptr.next
    print("None")
```

```python
def generate_llist():
    arr = range(1, 6)

    head = Node(arr[0])
    ptr = head

    for i in arr[1:]:
        node = Node(i)
        ptr.next = node
        ptr = node
    
    return head

head = generate_llist()
print_nodes(head)
```

```python
#   find element in linked_list return index, if not found return -1
#   https://www.geeksforgeeks.org/problems/search-in-linked-list-1664434326/1

def find_element(head, val):
    ptr = head
    
    if ptr is None:
        return -1
    
    index = 0
    while ptr is not None:
        if ptr.data == val:
            return index
        index += 1
        ptr = ptr.next
    
    if ptr is None:
        return -1
```

```python
head = generate_llist()
print_nodes(head)
find_element(head, 10)
```

```python
#   delete head of the linked_list
#   https://www.geeksforgeeks.org/problems/delete-head-of-linked-list/1

def delete_head(head):
    if head is None:
        return None
    return head.next

head = generate_llist()
print_nodes(head)
new_head = delete_head(head)
print_nodes(new_head)
```

```python
#   delete tail of linked_list
#   https://www.geeksforgeeks.org/problems/deletion-at-the-end-of-a-linked-list/1

def delete_tail(head):
    if head is None or head.next is None:
        return None
    
    ptr = head
    
    while ptr.next.next is not None:
        ptr = ptr.next
    
    ptr.next = None
    
    return head

head = generate_llist()
print_nodes(head)
new_head = delete_tail(head)
print_nodes(new_head)
```

```python
#   inserting at head of the linkedlist
#   https://www.geeksforgeeks.org/problems/linked-list-insertion-at-beginning/1

def insert_at_beg(head, x):
    new_node = Node(x)
    if head is not None:
        new_node.next = head
    
    return new_node

head = generate_llist()
print_nodes(head)
new_head = insert_at_beg(head, -1)
print_nodes(new_head)
```

```python
#   inserting node at last of linkedlist
#   https://www.geeksforgeeks.org/problems/linked-list-insertion-1587115620/1

def insert_at_last(head, x):
    new_node = Node(x)
    if head is None:
        return new_node
    
    ptr = head
    while ptr.next is not None:
        ptr = ptr.next
    
    ptr.next = new_node
    
    return head

head = generate_llist()
print_nodes(head)
new_head = insert_at_last(head, -1)
print_nodes(new_head)
```

```python
#   insert node in linkedlist
#   https://www.geeksforgeeks.org/problems/insertion-at-a-given-position-in-a-linked-list/1

def insert_node(head, index, x):
    new_node = Node(x)
    if index == 0:
        if head is not None:
            new_node.next = head
        return new_node
    
    curr = head.next
    prev = head
    calc_index = 1
    
    while curr is not None:
        if calc_index == index:
            new_node.next = curr
            prev.next = new_node
            break
        
        prev = curr
        curr = curr.next
        calc_index += 1
    
    if calc_index == index:
        prev.next = new_node
    
    return head

head = generate_llist()
print_nodes(head)
new_head = insert_node(head, 0, 100)
print_nodes(new_head)
```

```python
# doubly linkedlist

class DNode:
    def __init__(self, x):
        self.prev = None
        self.data = x
        self.next = None

arr2 = [1,2,3,4]

def create_llist(arr):
    if len(arr) == 0:
        return None
    head = DNode(arr[0])
    prev = head
    
    for i in arr[1:]:
        new_node = DNode(i)
        new_node.prev = prev
        prev.next = new_node
        
        prev = prev.next
    
    return head

def print_dll(head):
    ptr = head
    tail = None
    
    while ptr is not None:
        print(ptr.data, end="->")
        tail = ptr
        ptr = ptr.next
    print("None")
    ptr = tail
    while ptr is not None:
        print(ptr.data, end="->")
        ptr = ptr.prev
    print("None")

new_head = create_llist(arr2)
print_dll(new_head)
    
```

```python
#   https://www.geeksforgeeks.org/problems/delete-head-of-doubly-linked-list/1

def delete_head_dll(head):
    if head is None or head.prev is None and head.next is None:
        return None
    
    head = head.next
    head.prev = None
    
    return head

arr2 = [1,2,3]
head = create_llist(arr2)
print_dll(head)
new_head = delete_head_dll(head)
print_dll(new_head)
```

```python
#   https://www.geeksforgeeks.org/problems/delete-node-in-doubly-linked-list/1

def delete_tail_dll(head):
    if head is None or head.prev is None and head.next is None:
        return None
    
    curr = head
    
    while curr.next is not None:
        curr = curr.next
    
    curr.prev.next = None
    curr.prev = None
    del curr
    return head

arr2 = [4,5,6]
head = create_llist(arr2)
print_dll(head)
new_head = delete_tail_dll(head)
print_dll(new_head)
```

```python
#   reverse a dll
#   http://geeksforgeeks.org/problems/reverse-a-doubly-linked-list/1

def reverse(head):
    ptr = head
    new_head = head # working as prev node
    
    while ptr is not None:
        prev = ptr.prev
        next = ptr.next
        
        ptr.prev = next
        ptr.next = prev
        
        new_head = ptr
        ptr = ptr.prev
    
    return new_head
```

```python
#   odd-even linkedlist
#   http://leetcode.com/problems/odd-even-linked-list/

def oddEvenList(head):
    if head is None or head.next is None:
        return head
    
    odd = head
    even = head.next
    even_head = even

    while even is not None and even.next is not None:
        odd.next = odd.next.next
        odd = odd.next
        even.next = even.next.next
        even = even.next
    
    odd.next = even_head

    return head
```

```python
#   sort 0s, 1s and 2s in linkedlist
#   https://www.geeksforgeeks.org/problems/given-a-linked-list-of-0s-1s-and-2s-sort-it/1

def segregate(head):
    dummy_zero = Node(-1)
    zero = dummy_zero
    dummy_one = Node(-1)
    one = dummy_one
    dummy_two = Node(-1)
    two = dummy_two
    
    ptr = head
    
    while ptr is not None:
        if ptr.data == 0:
            zero.next = ptr
            zero = ptr
        elif ptr.data == 1:
            one.next = ptr
            one = ptr
        else:
            two.next = ptr
            two = ptr
        ptr = ptr.next
    
    if dummy_one.next is not None:
        zero.next = dummy_one.next
    else:
        zero.next = dummy_two.next
    
    one.next = dummy_two.next
    two.next = None
    
    return dummy_zero.next

```

```python
# check palindrome
# https://www.geeksforgeeks.org/problems/check-if-linked-list-is-pallindrome/1

'''
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

'''

class Solution:
    def isPalindrome(self, head):
        # code here
        slow = fast = head
        while fast.next is not None and fast.next.next is not None:
            slow = slow.next
            fast = fast.next.next
        
        rev_new_head = self.reverse(slow.next)
        ptr1 = head
        ptr2 = rev_new_head
        is_palindrome = True
        while ptr2 is not None:
            if ptr1.data != ptr2.data:
                is_palindrome = False
                break
            
            ptr1 = ptr1.next
            ptr2 = ptr2.next
        
        self.reverse(rev_new_head)
        
        return is_palindrome
    
    def reverse(self, head):
        if head is None or head.next is None:
            return head
        
        curr = head
        prev = next = None
        
        while curr is not None:
            next = curr.next
            curr.next = prev
            prev = curr
            curr = next
        
        return prev
        
```
